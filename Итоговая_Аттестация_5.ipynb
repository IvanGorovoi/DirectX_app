```python
import sqlite3
import pandas
import matplotlib.pyplot as plt
import plotly.express as px
```

Поделючаем базу данных

```python
conn = sqlite3.connect('lab5_new.db')
if(conn):print("База данных, успешно подключена!")
```

Считываем названия всех таблиц и формируем из них список.

```python
cur = conn.cursor()
#Считываем все существующие таблицы в Базе данных
table_list = [a for a in cur.execute("SELECT name FROM sqlite_master WHERE type = 'table'")]
string = ""
tables = []
#Заменяем тип данных с кортежа на список (для параметризорованных запросов к БД)
for i in range(len(table_list)):
  for j in range(len(table_list[i])):
    string+=table_list[i][j]
  tables.append(string)
  string = ""
```

```python
print('Выберите какую таблицу хотите отобразить: ')
for i in range(len(tables)):
  if(i!=0):print(f"{i}. {tables[i]}")
choice = int(input(":"))
pandas.read_sql_query("SELECT * FROM " + tables[choice],conn)
```

```python
string = ""
part1 = []

for i in range(len(columns)):
  if(columns[i]==','):
    part1.append(string)
    string = ""
  else: string+=columns[i]
  if(i==len(columns)-1): part1.append(string)
final_part1 = ""

for i in range(len(part1)):
  if(i==len(part1)-1):final_part1+=tables[choice]+"."+part1[i]
  else: final_part1+=tables[choice]+"."+part1[i]+", "

print(final_part1)

final_part2 = ""
if(columns1!= ""): final_part2+=", "
string = ""
part2 = []
for i in range(len(columns1)):
  if(columns1[i]==','):
    part1.append(string)
    string = ""
  else: string+=columns1[i]
  if(i==len(columns1)-1): part2.append(string)


for i in range(len(part2)):
  if(i==len(part2)-1):final_part2+=tables[choice1]+"."+part2[i]
  else: final_part2+=tables[choice1]+"."+part2[i]+", "
print(final_part2)
```
